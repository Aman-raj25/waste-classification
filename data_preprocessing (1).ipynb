```
import os
from PIL import Image

# Define input and output directories
input_dir = "/content/drive/MyDrive/Colab Notebooks/Machine Learning /project/Dataset/Plastic"  # This is where you put your original images
output_dir = "/content/resized_dataset/plastic"  # Processed images will be stored here

# Create output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Get all image files from input directory (excluding the output folder)
image_files = [f for f in os.listdir(input_dir)
               if f.lower().endswith(('png', 'jpg', 'jpeg', 'bmp', 'gif'))]

# Process images
for idx, filename in enumerate(sorted(image_files)):  # Sort ensures sequential numbering
    input_path = os.path.join(input_dir, filename)
    output_path = os.path.join(output_dir, f"PL-{idx:03d}.jpg")  # Ensure .jpg extension

    with Image.open(input_path) as img:
        # Convert to RGB (removes transparency from PNGs)
        img = img.convert("RGB")

        # Resize image to 299x299
        img = img.resize((224, 224))

        # Save image as JPG
        img.save(output_path, format="JPEG")

    print(f"Processed {filename} -> {output_path}")

print("All images have been processed and saved in dataset/output/!")
```
